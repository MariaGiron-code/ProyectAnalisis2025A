```python
import pandas as pd

def analizar_calidad_datos(df):
    print("--- Análisis de Calidad de Datos ---")
    
    # 1. Contar duplicados
    num_duplicados = df.duplicated().sum()
    print(f"\nNúmero de filas duplicadas (exactas): {num_duplicados}")
    
    if num_duplicados > 0:
        print("  -> Se encontraron duplicados que deben ser eliminados para un análisis correcto.")
    else:
        print("  -> ¡No se encontraron filas duplicadas! El DataFrame está limpio en este aspecto.")

    # 2. Contar valores nulos (vacíos)
    print("\n--- Conteo de Valores Nulos por Columna ---")
    
    # Sumar los valores nulos por columna
    valores_nulos = df.isnull().sum()
    
    # Calcular el porcentaje de valores nulos
    porcentaje_nulos = (valores_nulos / len(df)) * 100
    
    # Crear un DataFrame de resumen para los nulos
    df_nulos = pd.DataFrame({'Valores Nulos': valores_nulos, 'Porcentaje (%)': porcentaje_nulos})
    
    # Filtrar para mostrar solo las columnas con valores nulos
    df_nulos = df_nulos[df_nulos['Valores Nulos'] > 0].sort_values(by='Valores Nulos', ascending=False)
    
    if not df_nulos.empty:
        print(df_nulos)
        print("\n  -> Es importante decidir cómo manejar estos valores nulos (ej. eliminarlos, rellenarlos).")
    else:
        print("  -> ¡No se encontraron valores nulos en ninguna columna! El DataFrame está completo.")
    
#Ejecucio:
print(analizar_calidad_datos(event1))
print(analizar_calidad_datos(event2))
print(analizar_calidad_datos(event3))
print(analizar_calidad_datos(event4))
print(analizar_calidad_datos(event5))
print(analizar_calidad_datos(event6))
print(analizar_calidad_datos(event7))
```

```python
#Eliminar datos Vacios
def eliminar_datos_vacios(df, axis=0, how='any', subset=None):
    filas_iniciales, columnas_iniciales = df.shape
    df_limpio = df.dropna(axis=axis, how=how, subset=subset)
    filas_finales, columnas_finales = df_limpio.shape
    
    # Imprimir un resumen del cambio
    if axis == 0:
        print(f"Eliminando filas con valores nulos...")
        print(f"Filas eliminadas: {filas_iniciales - filas_finales}")
        print(f"DataFrame ahora tiene {filas_finales} filas y {columnas_finales} columnas.")
    else: # axis == 1
        print(f"Eliminando columnas con valores nulos...")
        print(f"Columnas eliminadas: {columnas_iniciales - columnas_finales}")
        print(f"DataFrame ahora tiene {filas_finales} filas y {columnas_finales} columnas.")
        
    return df_limpio
#Ejecucion
event1_clean=eliminar_datos_vacios(event1)
analizar_calidad_datos(event1_clean)

event2_clean=eliminar_datos_vacios(event2)
analizar_calidad_datos(event2_clean)

event3_clean=eliminar_datos_vacios(event3)
analizar_calidad_datos(event3_clean)

event4_clean=eliminar_datos_vacios(event4)
analizar_calidad_datos(event4_clean)

event5_clean=eliminar_datos_vacios(event5)
analizar_calidad_datos(event5_clean)

event6_clean=eliminar_datos_vacios(event6)
analizar_calidad_datos(event6_clean)

event7_clean=eliminar_datos_vacios(event7)
analizar_calidad_datos(event7_clean)
```

```python
#Creamos nuestra data set unificada para todo los equipos eouropeos contando champions, copa del mundo

#Selecionamos cuantos datos por eventos
eventEngland=event1_clean.head(20000)
eventChampions=event2_clean.head(20000)
eventFrance=event3_clean.head(20000)
eventGermany=event4_clean.head(20000)
eventItaly=event5_clean.head(20000)
eventSpain=event6_clean.head(20000)
eventWorld=event7_clean.head(20000)

#unimos todos los event
# 1. Crear una lista con todos los DataFrames que quieres combinar
lista_de_dataframes = [
    eventEngland,
    eventChampions,
    eventFrance,
    eventGermany,
    eventItaly,
    eventSpain,
    eventWorld
]

eventos_unidos = pd.concat(lista_de_dataframes, ignore_index=True)

# 3. Mostrar la información del DataFrame combinado
print("Información del DataFrame combinado:")
print(eventos_unidos.info())
```

```python
#Correcion de los datos eliminados correctamente
analizar_calidad_datos(eventos_unidos)
```

```python
#Eliminar columnas Innecesarias
# Columnas a eliminar
columnas_a_eliminar = ['eventId', 'tags', 'positions', 'subEventId']

# Función para eliminar columnas
def eliminarColumnas(dataframe, columnas_a_eliminar):
    return dataframe.drop(columns=[col for col in columnas_a_eliminar if col in dataframe.columns])

# Aplicar la función y guardar el resultado
eventos_final = eliminarColumnas(eventos_unidos, columnas_a_eliminar)
```
